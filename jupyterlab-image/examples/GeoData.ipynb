```python
import geopandas as gpd
import pandas as pd
import mplleaflet
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/de.csv')
df.head()
```

```python
max_pop = max(df['population'])
df['relative_pop'] = df['population']/max_pop
df.head()
```

```python

lats = [x for x in df['lat']]
lons = [y for y in df['lng']]
population = [r for r in df['relative_pop']]

fig = plt.figure(figsize=(20,20))
for i in range(len(lats)):
    if population[i]>0.8:
        plt.plot(lons[i], lats[i], 'ro', markersize=90*population[i])
    elif population[i]>0.4:
        plt.plot(lons[i], lats[i], 'bo', markersize=90*population[i])
    else:
        plt.plot(lons[i], lats[i], 'go', markersize=90*population[i])

mplleaflet.display(fig=fig)
```
